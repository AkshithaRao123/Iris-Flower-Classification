Importing necessary packages

```python
import pandas as pd
import numpy as np
```

Importing and exploring the dataset

```python
df = pd.read_csv('/content/drive/MyDrive/IrisClassification/Iris.csv')
df
```

```python
df.info()
```

Cleaning the dataset

```python
df = df.drop('Id', axis=1)
df
```

```python
species = set(list(df['Species']))
print(species)
```

```python
targetid = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
key_list = list(targetid.keys())
val_list = list(targetid.values())
```

```python
lst = []

for i in range(len(df['Species'])):
  s = df['Species'][i]
  lst.append(targetid[s])

df['target'] = lst
```

```python
df
```

Assigning features and labels

```python
x = df.iloc[:, :-2].values
y = df.iloc[:, -1].values
x
```

```python
y
```

Splitting into train and test sets

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.25, random_state = 40)
```

```python
dct = {'0':0, '1':0, '2':0}

for i in ytrain:
  if i==0:
    dct['0']+=1
  if i==1:
    dct['1']+=1
  if i==2:
    dct['2']+=1

dct
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit_transform(xtrain)
scaler.transform(xtest)
```

Training the model

#### Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
obj = LogisticRegression()

obj.fit(xtrain,ytrain)
```

```python
ypred = obj.predict(xtest)
```

```python
print(ypred)

predictions = pd.DataFrame(list(zip(xtest, ypred, ytest)), columns=['X','Predicted','Actual'])
```

```python
def match(val):
  pos = val_list.index(val)
  v = key_list[pos]
  return v
```

```python
predictions
```

Evaluating the model

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(ytest,ypred))
```

```python
from sklearn.metrics import recall_score,accuracy_score
```

```python
score = recall_score(ytest,ypred, average='weighted')
print("Recall score: ", score)
```

```python
score = accuracy_score(ytest, ypred)
print("Accuracy score: ", score)
```
